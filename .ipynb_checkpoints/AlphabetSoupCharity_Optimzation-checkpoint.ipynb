# **Optimization Attempt #1: Checking Features**



```python
# Create a copy of the dataframe for the optimization #1
application_df_opt1 = application_df.copy()
```

```python
# Separate categorical and numeric features for a closer look
cat_features = application_df.dtypes[application_df.dtypes == 'object'].index.to_list()
print(cat_features)
print(application_df[cat_features].nunique())

num_features = application_df.dtypes[application_df.dtypes != 'object'].index.to_list()
print(num_features)
```

```python
# Get value counts for STATUS column for closer look -- if interpretation is correct, this indicates if application is active or not
application_df_opt1['STATUS'].value_counts()
```

```python
# Since almost all of the applications have STATUS = 1, it is likely safe to drop the column
application_df_opt1 = application_df_opt1.drop(columns=['STATUS'])
application_df_opt1
```

```python
# Look at APPLICATION_TYPE value counts for binning
appType_vCounts = application_df_opt1['APPLICATION_TYPE'].value_counts()
appType_vCounts
```

```python
# ** Change cutoff value from 500 to 1000 (leave top 5 application types with Others)
# Create a list of application types to be replaced; use the variable name `application_types_to_replace`
appType_cutoff = 1000

appType_vCounts_df = pd.DataFrame(appType_vCounts).reset_index()
appType_vCounts_df.columns = ['application_type', 'value_count']
#display(appType_vCounts_df)

application_types_to_replace = appType_vCounts_df.loc[appType_vCounts_df['value_count']<appType_cutoff, 'application_type'].to_list()
#print(application_types_to_replace)

# Replace target application type data in dataframe
for app in application_types_to_replace:
    application_df_opt1['APPLICATION_TYPE'] = application_df_opt1['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df_opt1['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
appCls_vCounts = application_df_opt1['CLASSIFICATION'].value_counts()
#appCls_vCounts

# You may find it helpful to look at CLASSIFICATION value counts >1
appCls_vCounts.loc[appCls_vCounts >1]
```

```python
# ** Keep cutoff value at 1000 (leave top 5 classifications with Others)
# Create a list of classifications to be replaced; use the variable name `classifications_to_replace`
appCls_cutoff = 1000

appCls_vCounts_df = pd.DataFrame(appCls_vCounts).reset_index()
appCls_vCounts_df.columns = ['classification', 'value_count']
#display(appCls_vCounts_df)

classifications_to_replace = appCls_vCounts_df.loc[appCls_vCounts_df['value_count']<appCls_cutoff, 'classification'].to_list()
#print(classifications_to_replace)

# Replace target classification data in reduced features dataframe
for cls in classifications_to_replace:
    application_df_opt1['CLASSIFICATION'] = application_df_opt1['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df_opt1['CLASSIFICATION'].value_counts()
```

```python
# Look at INCOME_AMT value counts for possible binning
incomeAmt_vCounts = application_df['INCOME_AMT'].value_counts()
incomeAmt_vCounts
```

```python
# ** Creat new bins for incomes 1-24999 and 1M-50M+
# Create a lists of income amounts to be replaced
under25K = ['1-9999', '10000-24999']
over1M = ['1M-5M', '10M-50M', '5M-10M', '50M+']

# Replace target income data into new bins
for income in under25K:
  application_df_opt1['INCOME_AMT'] = application_df_opt1['INCOME_AMT'].replace(income,"1-24999")

for income in over1M:
  application_df_opt1['INCOME_AMT'] = application_df_opt1['INCOME_AMT'].replace(income,"1M-50M+")

# Check to make sure binning was successful
application_df_opt1['INCOME_AMT'].value_counts()
```

```python
# Look at AFFILIATION value counts for possible binning
aff_vCounts = application_df['AFFILIATION'].value_counts()
aff_vCounts
```

```python
# ** Combine Family/Parent, National, Regional, and Other into same bin
# Create a list of affiliations replaced; use the variable name `aff_to_replace`
aff_to_replace = ['Family/Parent', 'National', 'Regional', 'Other']

# Replace target affiliation data into new bin
for aff in aff_to_replace:
  application_df_opt1['AFFILIATION'] = application_df_opt1['AFFILIATION'].replace(aff,"Other")

# Check to make sure binning was successful
application_df_opt1['AFFILIATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_df_opt1_n = pd.get_dummies(application_df_opt1)
application_df_opt1_n
```

```python
# Split our preprocessed data into our features and target arrays
X = application_df_opt1_n.drop(columns=['IS_SUCCESSFUL'])
y = application_df_opt1_n['IS_SUCCESSFUL']

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model 
# ** Keep original number of input features and hidden nodes for each layer.
inputFeatures_n = len(X_train_scaled[0])
hiddenNodes_L1 = 80
hiddenNodes_L2 = 30

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hiddenNodes_L1, activation='relu', input_dim=inputFeatures_n))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hiddenNodes_L2, activation='relu'))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Check the structure of the model
nn.summary()

# Compile the model
nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
# ** Keep number of epochs
cp = tf.keras.callbacks.ModelCheckpoint('/content/drive/MyDrive/BOOTCAMP/Deep-Learning-Challenge/Output/weights_opt1.h5', save_weights_only=True, periods=5)
fit_model = nn.fit(X_train_scaled, y_train, epochs=100, callbacks=[cp])
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled, y_test, verbose=2)
print(f'Original Results -- Loss: 0.5729537606239319, Accuracy: 0.729912519454956')
print(f"New Results ------- Loss: {model_loss}, Accuracy: {model_accuracy}")

# Export our model to HDF5 file -- name it AlphabetSoupCharity_Optimization_a1.h5
nn.save('/content/drive/MyDrive/BOOTCAMP/Deep-Learning-Challenge/Output/AlphabetSoupCharity_Optimization_a1.h5')
```

## Results 
The Loss is slightly lower and Accuracy slightly higher than the results from the original model. However, we have not yet reached the target of 75% accuracy. Based on the improved results, I will keep the changes made during this attempt for the next round.

# **Optimization Attempt #2: Altering Number of Hidden Layers and Nodes**

```python
# ** Compile, Train, and Evaluate the model, keeping the changes from attempt 1
# ** Add a hidden layer, increase number of nodes in each layer
inputFeatures_n = len(X_train_scaled[0])
hiddenNodes_L1 = 250
hiddenNodes_L2 = 125
hiddenNodes_L3 = 75

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hiddenNodes_L1, activation='relu', input_dim=inputFeatures_n))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hiddenNodes_L2, activation='relu'))

# Third hidden layer
nn.add(tf.keras.layers.Dense(units=hiddenNodes_L3, activation='relu'))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Check the structure of the model
nn.summary()

# Compile the model
nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
# ** Keep number of epochs
cp = tf.keras.callbacks.ModelCheckpoint('/content/drive/MyDrive/BOOTCAMP/Deep-Learning-Challenge/Output/weights_opt2.h5', save_weights_only=True, periods=5)
fit_model = nn.fit(X_train_scaled, y_train, epochs=100, callbacks=[cp])
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled, y_test, verbose=2)
print(f'Original Results -- Loss: 0.5729537606239319, Accuracy: 0.729912519454956')
print(f'Attempt 1 --------- Loss: 0.559613049030304, Accuracy: 0.7303789854049683')
print(f"New Results ------- Loss: {model_loss}, Accuracy: {model_accuracy}")

# Export our model to HDF5 file -- name it AlphabetSoupCharity_Optimization_a2.h5
nn.save('/content/drive/MyDrive/BOOTCAMP/Deep-Learning-Challenge/Output/AlphabetSoupCharity_Optimization_a2.h5')
```

## Results 
The Loss increased once again, though the Accuracy is also marginally higher than the results from the previous attempt. Based on the improved results, I will keep the changes made during this attempt for the next round.

# **Optimization Attempt #3: Altering Activation Functions**

```python
# ** Compile, Train, and Evaluate the model, keeping the changes from attempt 1 and 2
# ** Change activation function for hidden layers from relu to tanh
inputFeatures_n = len(X_train_scaled[0])
hiddenNodes_L1 = 250
hiddenNodes_L2 = 125
hiddenNodes_L3 = 75

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hiddenNodes_L1, activation='tanh', input_dim=inputFeatures_n))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hiddenNodes_L2, activation='tanh'))

# Third hidden layer
nn.add(tf.keras.layers.Dense(units=hiddenNodes_L3, activation='tanh'))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Check the structure of the model
nn.summary()

# Compile the model
nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
# ** Keep number of epochs -- attempts with 50 and 20 epochs resulted in lower accuracy scores
cp = tf.keras.callbacks.ModelCheckpoint('/content/drive/MyDrive/BOOTCAMP/Deep-Learning-Challenge/Output/weights_opt3.h5', save_weights_only=True, periods=5)
fit_model = nn.fit(X_train_scaled, y_train, epochs=100, callbacks=[cp])
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled, y_test, verbose=2)
print(f'Original Results -- Loss: 0.5729537606239319, Accuracy: 0.729912519454956')
print(f'Attempt 1 --------- Loss: 0.559613049030304, Accuracy: 0.7303789854049683')
print(f'Attempt 2 --------- Loss: 0.578855037689209, Accuracy: 0.7307288646697998')
print(f"New Results ------- Loss: {model_loss}, Accuracy: {model_accuracy}")

# Export our model to HDF5 file -- name it AlphabetSoupCharity_Optimization_a3.h5
nn.save('/content/drive/MyDrive/BOOTCAMP/Deep-Learning-Challenge/Output/AlphabetSoupCharity_Optimization_a3.h5')
```
